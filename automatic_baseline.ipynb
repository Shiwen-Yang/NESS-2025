```python
# 1. Import library
from autogluon.tabular import TabularPredictor
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
# 2. Load the data
train_df = pd.read_csv('Data/original/train_2025.csv')    # Adjust path if needed
test_df = pd.read_csv('Data/original/test_2025.csv')      # Adjust path if needed
train_df, val_df = train_test_split(train_df, test_size=0.2, stratify=train_df["fraud"], random_state=42)
```

```python
predictor_0 = TabularPredictor(
    label="fraud",
    eval_metric="f1",
    problem_type="binary"
).fit(
    train_data=train_df,
    holdout_frac=0.2,
    presets="best"
)

predictor_0.evaluate(val_df)
```

```python
predictor_0.leaderboard(val_df)
```

```python
# 4. Predict on the test set
predictions = predictor_0.predict(test_df)

# 5. Save predictions to CSV
submission = pd.DataFrame({
    "claim_number": test_df["claim_number"],  # Important: use the original claim_number
    "fraud": predictions                      # Your predicted fraud labels (0 or 1)
})
submission.to_csv("submissions/submission_autogluon_baseline.csv", index=False)
```
