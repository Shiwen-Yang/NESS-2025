```python
import pandas as pd
import numpy as np

train_df = pd.read_csv('Original_Data/train_2025.csv')
info_0 = pd.read_csv('Create_Tune_Models/logs/subset_info_3.csv')
info_1 = pd.read_csv('Create_Tune_Models/logs/subset_info_4.csv')
info_2 = pd.read_csv('Create_Tune_Models/logs/subset_info_5.csv')


def filter_presence_info(presence_info_df, difference_min=0.06, info_min=0.03):
    """
    Filters the presence_info_df based on thresholds for 'difference' and 'info'.

    Parameters:
    ----------
    presence_info_df : pd.DataFrame
        The DataFrame containing at least 'difference' and 'info' columns.
    difference_min : float, optional
        Minimum absolute value of 'difference' required to keep a row (default 0.06).
    info_min : float, optional
        Minimum value of 'info' required to keep a row (default 0.03).

    Returns:
    -------
    pd.DataFrame
        Filtered DataFrame.
    """
    filtered_df = presence_info_df[
        (np.abs(presence_info_df['difference']) > difference_min) &
        (presence_info_df['info'] > info_min)
    ]
    return filtered_df
```

```python
f_info_0 = filter_presence_info(info_0, difference_min=0.05, info_min=0.03)
```

```python
f_info_1 = filter_presence_info(info_1, difference_min=0.06, info_min=0.035)

f_info_1.sort_values(by = 'info', ascending=False)
```

```python
# Convert feature columns to sets
features_1 = set(f_info_0['feature'])
features_2 = set(f_info_1['feature'])

# Find the intersection
common_features = features_1.intersection(features_2)

# Count how many features overlap
print(f"Number of overlapping features: {len(common_features)}")

# (Optional) See the actual overlapping feature names
print("Overlapping features:", common_features)

```
